```python
def check(nums):
    nw = []
    if len(nums) == 0:
        return 0
    if len(nums) == 1:
        return 1
    if len(nums) == 2:
        if nums[0] == nums[0]:
            return 1
        return 2
    for idx in range(1, len(nums)):
        sym = nums[idx - 1] - nums[idx]
        if sym > 0:
            current = 1
        elif sym == 0:
            current = 0
            continue
        else:
            current = -1
        if not nw:
            nw.append(current)
        if current != nw[-1]:
            nw.append(current)
    return len(nw) + 1
for inpt, output in tests:
    ot = check(inpt)
    assert ot == output, (inpt, ot, output)
```
