```python
tests=[("9,3,4,#,#,1,#,#,2,#,6,#,#",True),
       ("1,#", False),
       ("9,#,#,1", False),
       ("9,#,92,#,#", True)
      ]
class Solution(object):
    def to_list(self, inpt):
        ls = inpt.split(',')
#         _ls = []
#         for item in ls:
#             if item.isdigit():
#                 _ls.append(int(item))
#             else:
#                 _ls.append(item)
        return ls
    def isValidSerialization(self, preorder):
        """
        :type preorder: str
        :rtype: bool
        """
        ls = preorder.split(',')
        deep = 0
        for item in ls:
            if deep == -1:
                return False
            if item == "#":
                deep -= 1
            else:
                deep += 1
        return deep == -1
    
sl = Solution()
for inpt, res in tests:
    result = sl.isValidSerialization(inpt)
    assert result == res , (inpt, res, result)
```
